```python
import numpy as np
import matplotlib.pyplot as plt
names = np.array(["Rohan","Nandini","Bhawani","Sabnam","Deepak"])
age = np.array([21,22,23,22,21])
arr = np.random.uniform(0,20,10)
# plt.xlabel("Names")
# plt.ylabel("Age")
# plt.title("Persons Data")
plt.subplot(2,1,1)
plt.plot(arr)
plt.subplot(2,1,2)
plt.plot(names,age,mfc="violet",c="brown",marker="o")
plt.title("Age Data")
plt.xlabel("Names")
plt.ylabel("Age")
plt.show()
```
