```python
import pandas as pd
file = pd.read_csv("real.csv")
data = pd.DataFrame(file)
print(data)
data.loc[0:8:2,"Country"] = None
data.loc[9,"InvoiceNo"] = None
print(data)
# steps to clean the data
#1) checking the null value
print(data.isnull())
print(data.count())
# 2) correcting the missing values
# a) fill the missing values to 0 or input your value
data["Country"] = data["Country"].fillna("India")
print(data["Country"])
# b) fill the mean value
mean1 = data["InvoiceNo"].mean()
data["InvoiceNo"] = data["InvoiceNo"].fillna(mean1)
print(data["InvoiceNo"])
# c) pad and bfill method
# => pad means forwards
data["InvoiceNo"] = data["InvoiceNo"].fillna(method="Pad")
print(data["InvoiceNo"])
data["InvoiceNo"] = data["InvoiceNo"].fillna(method="bfill")
print(data["InvoiceNo"])
```
