```python
import pandas as pd
import numpy as np
file = pd.read_csv("real.csv")
data = pd.DataFrame(file)
data.loc[3:16:2,"InvoiceNo"] = None
print(data)
```

```python
null_val = data.isnull()
print(null_val)
```

```python
mean1 = data["InvoiceNo"].mean()
data["InvoiceNo"] = data["InvoiceNo"].fillna(mean1)
data["Description"] = data["Description"].str.lower()
print(data)
print(data.duplicated())
data = data.drop_duplicates()
print(data)
data["new_column"] = np.arange(1,18)
data.rename(columns={"Description":"Product_Description"},inplace=True)
print(data)
```
