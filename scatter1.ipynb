```python
import matplotlib.pyplot as plt
import numpy as np
xcor = np.random.uniform(10,50,10)
ycor = np.random.uniform(10,50,10)
plt.scatter(xcor, ycor,color="red",s=140)
#
xcor = np.random.uniform(10,50,10)
ycor = np.random.uniform(10,50,10)
color = np.arange(10,101,10)
plt.scatter(xcor, ycor,c=color,cmap="Oranges")
plt.colorbar()
plt.show()
x = np.arange(1,11)
plt.bar(x,ycor,color="green",width=0.4)
plt.show()
```
